# Experiment 5: Expert System
**Objective:** Build a rule-based Expert System for Employee Performance Evaluation.

**Problem Statement:** # An expert system mimics the decision-making ability of a human expert. This system uses a Knowledge Base (rules about performance) and an Inference Engine (logic to apply rules) to evaluate employees.

**Logic:** The system asks for scores (1-10) in various categories and applies conditional logic to provide a final performance grade.

```python
class PerformanceExpertSystem:
    def __init__(self):
        self.rules = []
        self.kpis = {
            "productivity": 0,
            "teamwork": 0,
            "punctuality": 0,
            "leadership": 0
        }

    def get_user_input(self):
        print("--- Employee Performance Evaluation System ---")
        print("Please rate the employee in the following categories (1-10):")
        for kpi in self.kpis:
            while True:
                try:
                    score = float(input(f"Enter score for {kpi.capitalize()}: "))
                    if 1 <= score <= 10:
                        self.kpis[kpi] = score
                        break
                    else:
                        print("Please enter a value between 1 and 10.")
                except ValueError:
                    print("Invalid input. Please enter a number.")

    def evaluate(self):
        avg_score = sum(self.kpis.values()) / len(self.kpis)
        p = self.kpis["productivity"]
        t = self.kpis["teamwork"]
        
        print("\n--- Evaluation Result ---")
        print(f"Average Score: {avg_score:.2f}")
        
        # Inference Engine (Rule-based logic)
        if avg_score >= 9 or (p >= 9 and t >= 8):
            return "Rating: Outstanding. Recommendation: Eligible for Promotion and Bonus."
        
        elif avg_score >= 7:
            return "Rating: Satisfactory. Recommendation: Standard Annual Increment."
        
        elif avg_score >= 5:
            return "Rating: Needs Improvement. Recommendation: Enroll in Performance Improvement Plan (PIP)."
        
        else:
            return "Rating: Poor. Recommendation: Immediate review required; potential disciplinary action."

if __name__ == "__main__":
    expert = PerformanceExpertSystem()
    expert.get_user_input()
    result = expert.evaluate()
    print(result)
```
