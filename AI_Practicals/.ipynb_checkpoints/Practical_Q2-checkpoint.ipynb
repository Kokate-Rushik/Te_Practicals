# Experiment 2: A* Search Algorithm
**Objective:** Implement the A* algorithm for the 8-Puzzle problem.

**Problem Statement:** # The 8-puzzle consists of a 3x3 board with 8 numbered tiles and one blank space. The goal is to reach a specific target configuration using the minimum number of moves.

**Heuristic Used:** Manhattan Distance (sum of the absolute differences of the coordinates of the tiles from their goal positions).

```python
import heapq

class Node:
    def __init__(self, state, parent=None, move="", g=0, h=0):
        self.state = state
        self.parent = parent
        self.move = move
        self.g = g  # Cost from start to current node
        self.h = h  # Heuristic cost to goal
        self.f = g + h

    def __lt__(self, other):
        return self.f < other.f

def get_manhattan_distance(state, goal):
    distance = 0
    for i in range(1, 9):
        curr_pos = state.index(i)
        goal_pos = goal.index(i)
        curr_coords = (curr_pos // 3, curr_pos % 3)
        goal_coords = (goal_pos // 3, goal_pos % 3)
        distance += abs(curr_coords[0] - goal_coords[0]) + abs(curr_coords[1] - goal_coords[1])
    return distance

def get_neighbors(state):
    neighbors = []
    blank_idx = state.index(0)
    r, c = blank_idx // 3, blank_idx % 3
    moves = {"Up": (-1, 0), "Down": (1, 0), "Left": (0, -1), "Right": (0, 1)}

    for move, (dr, dc) in moves.items():
        nr, nc = r + dr, c + dc
        if 0 <= nr < 3 and 0 <= nc < 3:
            new_state = list(state)
            target_idx = nr * 3 + nc
            new_state[blank_idx], new_state[target_idx] = new_state[target_idx], new_state[blank_idx]
            neighbors.append((tuple(new_state), move))
    return neighbors

def solve_8_puzzle(start_state, goal_state):
    open_list = []
    start_node = Node(start_state, None, "", 0, get_manhattan_distance(start_state, goal_state))
    heapq.heappush(open_list, start_node)
    
    closed_list = set()
    
    while open_list:
        current_node = heapq.heappop(open_list)
        
        if current_node.state == goal_state:
            path = []
            while current_node.parent:
                path.append((current_node.move, current_node.state))
                current_node = current_node.parent
            return path[::-1]
        
        closed_list.add(current_node.state)
        
        for next_state, move in get_neighbors(current_node.state):
            if next_state in closed_list:
                continue
                
            g = current_node.g + 1
            h = get_manhattan_distance(next_state, goal_state)
            neighbor_node = Node(next_state, current_node, move, g, h)
            heapq.heappush(open_list, neighbor_node)
            
    return None

# Initial state (0 represents the blank space)
initial = (1, 2, 3, 0, 4, 6, 7, 5, 8)
goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)

print("Starting A* Search...")
solution = solve_8_puzzle(initial, goal)

if solution:
    print(f"Goal reached in {len(solution)} moves!")
    for move, state in solution:
        print(f"Move: {move}")
        for i in range(0, 9, 3):
            print(state[i:i+3])
        print("-" * 10)
else:
    print("No solution found.")
```
