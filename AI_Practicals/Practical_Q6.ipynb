# Experiment 6: Rule-Based Chatbot
**Objective:** Implement a Chatbot using rule-based logic (if-else statements) to solve customer login issues.

**Problem Statement:** # Customer support often deals with repetitive login queries. This chatbot acts as the first line of support, diagnosing whether the issue is a forgotten password, account lockout, or technical glitch.

**Implementation:** A simple interactive loop that parses keywords from user input to provide solutions.

```python
def login_chatbot():
    print("--- Customer Support Chatbot: Login Issues ---")
    print("AI: Hello! I am your Login Assistant. How can I help you today?")
    print("(Type 'exit' or 'quit' to stop)\n")

    while True:
        user_input = input("User: ").lower()

        # Exit condition
        if user_input == 'exit' or user_input == 'quit':
            print("AI: Thank you for visiting. Have a great day!")
            break

        # Decision Tree using If-Else logic
        if "forgot" in user_input and "password" in user_input:
            print("AI: No problem! To reset your password, click 'Forgot Password' on the login page and check your email for a reset link.")
        
        elif "locked" in user_input or "lockout" in user_input:
            print("AI: Accounts are usually locked after 3 failed attempts for security. Please wait 30 minutes or verify your identity via SMS to unlock it.")
        
        elif "username" in user_input:
            print("AI: If you've forgotten your username, try using your registered email address instead.")
        
        elif "captcha" in user_input or "verification code" in user_input:
            print("AI: If the captcha isn't loading, try refreshing the page or clearing your browser cache.")
        
        elif "error" in user_input or "page not found" in user_input:
            print("AI: This might be a technical glitch. Please try opening the site in an Incognito/Private window or use a different browser.")
        
        elif "thank" in user_input:
            print("AI: You're very welcome! Is there anything else I can help you with?")
        
        elif "hello" in user_input or "hi" in user_input:
            print("AI: Hi there! Please describe the login problem you are facing.")
            
        else:
            print("AI: I'm not sure I understand that specific issue. Would you like me to connect you with a human agent?")

if __name__ == "__main__":
    login_chatbot()
```
