# **Experiment 8: Data Visualization I**
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to
see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (fare) varies with respect to passenger class and survival.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

print(f"Dataset Shape: {df.shape}")
print(df.head())
```

### 1. Pattern Recognition
We'll start by checking the survival rate based on Gender and Class. 
These are usually the strongest boosters for survival probability.

```python
# Pattern 1: Survival rate by Sex
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', hue='sex', data=df, palette='Set1')
plt.title('Survival Count by Gender')
plt.show()

# Pattern 2: Survival rate by Class
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', hue='pclass', data=df, palette='viridis')
plt.title('Survival Count by Passenger Class')
plt.show()
```

### 2. Ticket Price (Fare) Analysis
Checking the distribution of the 'fare' variable and how it relates to passenger class.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['fare'], bins=40, kde=True, color='blue')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(hue='pclass', y='fare', data=df, palette='Set2', legend=False)
plt.title('Ticket Price variation by Passenger Class')
plt.ylim(0, 300)
plt.show()
```

### Correlation Heatmap
To find numeric subtasks that impact survival, we look at the correlation matrix.

```python
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Numeric Feature Correlation')
plt.show()
```
