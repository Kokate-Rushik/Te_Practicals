## 1. Loading datasets

```python
import pandas as pd

iris_data = pd.read_csv("datasets/Iris.csv")

print(iris_data.head())
```

## 2. Summary statistics grouped by Species

```python
grouped_stats = iris_data.groupby('Species')['SepalLengthCm'].agg(['mean', 'median', 'min', 'max', 'std'])

print("### Summary Statistics of Sepal Length Grouped by Species ###")
print(grouped_stats)

# Creating a list containing numeric values for each response (Mean Sepal Lengths)
mean_list = grouped_stats['mean'].tolist()
print(f"\nList of Mean Sepal Lengths: {mean_list}")
```

## 3. Detailed Statistics for Specific Species

```python
def get_detailed_stats(df, species_column, species_name):
    # This check ensures we match even if 'Iris-' is present
    subset = df[df[species_column].str.contains(species_name, case=False)]
    
    if not subset.empty:
        print(f"\n--- Statistical Details for {species_name} ---")
        print(subset.describe())
    else:
        print(f"\nNo data found for {species_name}. Check spelling!")

# 4. Display species stats
for s in ['setosa', 'versicolor', 'virginica']:
    get_detailed_stats(iris_data, 'Species', s)
```
