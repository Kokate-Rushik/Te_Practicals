# **Experiment 9: Data Visualization II**
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution 
of age with respect to each gender along with the information about whether they survived 
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

plot_data = df.dropna(subset=['age'])
```

### 1. Box Plot: Age Distribution by Gender and Survival
This plot will show the quartiles of age, the median, and any potential outliers.

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='sex', y='age', hue='survived', data=plot_data, palette='Set1')

plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender (Sex)')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

### 2. Observations and Inferences
Based on the generated box plot, we can draw the following statistical inferences:

```python
summary = plot_data.groupby(['sex', 'survived'])['age'].describe()
print("Summary Statistics for Age:")
print(summary)
```
